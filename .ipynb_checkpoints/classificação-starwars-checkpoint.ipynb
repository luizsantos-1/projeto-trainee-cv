```python
plt.imshow(img)
plt.show()
```

```python
# rotação 180 com numpy
flipped=np.empty(np.shape(img))
linhas=np.shape(img)[0]-1
colunas=np.shape(img)[1]-1

for i in range (linhas):
    for j in range (colunas):
        flipped[linhas-i,colunas-j]=img[i,j]
for j in range (colunas):
    for i in range (linhas):
        flipped[linhas-i,colunas-j]=img[i,j]

flipped = np.array(flipped, dtype=np.uint8)
plt.imshow(flipped)
plt.show()
```

```python
# flip (direita-esquerda)
def flip_left_to_right(img):
    flipped=np.empty(np.shape(img))
    linhas=np.shape(img)[0]-1
    colunas=np.shape(img)[1]-1
    for i in range (linhas):
        for j in range (colunas):
            flipped [i][colunas-j]=img[i][j]
    flipped = np.array(flipped, dtype=np.uint8)
    return(flipped)
img = flip_left_to_right(img)
```

```python
# implementação brilho e contraste
alpha = np.random.random()/2 + 0.75 #contraste
beta = np.random.random()*20 #brilho
new_img = np.empty(np.shape(img))
for y in range(img.shape[0]):
    for x in range(img.shape[1]):
        for c in range(img.shape[2]):
            new_img[y,x,c] = np.clip(alpha*img[y,x,c] + beta, 0, 255)
new_img = np.array(new_img, dtype=np.uint8)
plt.imshow(new_img)
plt.show()
```

```python
#rotação, brilho, zoom aleatórios  funciona
def random_changes(img):
    linhas = np.shape(img)[0]
    colunas = np.shape(img)[1]

    center= (np.random.randint(linhas*(24/50),linhas*(26/50)),
             np.random.randint(colunas*(24/50),colunas*(26/50)))
    angle= (np.random.random()*360)
    scale= (np.random.random()/10)+0.95
    M = cv2.getRotationMatrix2D(center,angle,scale)
    img_rot = cv2.warpAffine(img,M,(colunas,linhas))
    return (img_rot)
#plt.imshow(img_rot)
#plt.show()
```

```python
import glob
import cv2
import numpy as np
import os 
```

```python
images_darth = np.array([cv2.imread(file) for file in glob.glob("/Users/luizeduardosantos/Documents/Turing/projeto-trainee-cv/Dataset Trainee Oficial Top/Darth Vader/*")])
images_yoda = np.array([cv2.imread(file) for file in glob.glob("/Users/luizeduardosantos/Documents/Turing/projeto-trainee-cv/Dataset Trainee Oficial Top/Yoda/*")])
images_trooper = np.array([cv2.imread(file) for file in glob.glob("/Users/luizeduardosantos/Documents/Turing/projeto-trainee-cv/Dataset Trainee Oficial Top/Stormtrooper/*")])
```

```python
def train_test_validation_split(arr_images, personagem):
    
    nimages = arr_images.shape[0]
    
    arr = np.arange(nimages)
    np.random.shuffle(arr)
    
    for i in range (nimages):
        if i <= nimages*0.8:
            directory = '/Users/luizeduardosantos/Documents/Turing/projeto-trainee-cv/train/'+ personagem
            os.chdir(directory)
            a = arr[i]
            image = arr_images[a]
        
            filename = 'train'+ str(i)
            cv2.imwrite(filename+'.jpeg',image)
            
            
        elif i > nimages*0.8 and i <= nimages*0.9:
            directory = '/Users/luizeduardosantos/Documents/Turing/projeto-trainee-cv/validation/'+personagem
            os.chdir(directory)
            a = arr[i]
            image = arr_images[a]
            
            filename = 'val'+ str(i)
            cv2.imwrite(filename+'.jpeg',image)
        else: 
            directory = '/Users/luizeduardosantos/Documents/Turing/projeto-trainee-cv/test/'+personagem
            os.chdir(directory)
            a = arr[i]
            image = arr_images[a]
        
            filename = 'test'+ str(i)
            cv2.imwrite(filename+'.jpeg',image)        
# talvez seja bom juntar as pastas de treino e validação porque o keras pode fazr essa divisão.
```

```python
train_test_validation_split(images_darth,'Darth Vader')
train_test_validation_split(images_trooper,'Stormtrooper')
train_test_validation_split(images_yoda, 'Yoda')
```

creating the array of training labels 

```python
train_images = np.concatenate((train_images_darth,
                               train_images_yoda,
                               train_images_trooper))
```

```python
val_images = np.concatenate((val_images_darth,
                               val_images_yoda,
                               val_images_trooper))
```

```python
n_images = train_images.shape[0]
training_labels = np.zeros((n_images,3), dtype=int)
d = train_images_darth.shape[0]
y = train_images_yoda.shape[0]
t = train_images_trooper.shape[0]
for i in range (d):
    training_labels[i][0] = 1
for i in range (y):
    training_labels[d + i][1] = 1
for i in range (t):
    training_labels[d + y +i][2] = 1
training_labels
```

```python
n_images = val_images.shape[0]
val_labels = np.zeros((n_images,3), dtype=int)
d = val_images_darth.shape[0]
y = val_images_yoda.shape[0]
t = val_images_trooper.shape[0]
for i in range (d):
    val_labels[i][0] = 1
for i in range (y):
    val_labels[d + i][1] = 1
for i in range (t):
    val_labels[d + y +i][2] = 1
```

resizing and turning the images into vectors

```python
n_images = train_images.shape[0]
resized_train_images = np.empty((n_images,408,408,3), dtype=int)

for i in range (n_images):
    resized_train_images[i] = cv2.resize(train_images[i],(408,408))

resized_train_images.shape         
```

```python
n_images = val_images.shape[0]
resized_val_images = np.empty((n_images,408,408,3), dtype=int)

for i in range (n_images):
    resized_val_images[i] = cv2.resize(val_images[i],(408,408))

resized_val_images.shape       
```

```python
vector_train_images = np.resize(resized_train_images, (1464,408*408*3))
vector_train_images.shape
```

```python
vector_val_images = np.resize(resized_val_images, (183,408*408*3))
vector_val_images.shape
```

```python
#setting 
```

```python
vector_train_images = vector_train_images/255
```

```python
vector_val_images = vector_val_images/255
```

```python
model = keras.Sequential()
model.add(keras.layers.Dense(1024,activation='relu',input_shape = (408*408*3,)))
model.add(keras.layers.Dense(256, activation= 'relu'))
model.add(keras.layers.Dense(3, activation= 'softmax'))
```

```python
model.compile(optimizer='SGD',
              loss = tf.keras.losses.CategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(vector_train_images,
          training_labels,
          batch_size=4, 
          epochs=3, 
          validation_data = (vector_val_images,val_labels)
        )
         
```

```python
model.predict(vector_val_images)
```

```python
idg_train = ImageDataGenerator(rescale=1./255,
                         rotation_range=45,
                         width_shift_range=.15,
                         height_shift_range=.15,
                         horizontal_flip=True,
                         zoom_range=0.5
                        )
```

```python
idg_validation_test = ImageDataGenerator(rescale=1./255)
```

```python
train_dir = '/Users/luizeduardosantos/Documents/Turing/projeto-trainee-cv/train'
batch_size = 1 #arrumar

train_images = idg_train.flow_from_directory(batch_size=batch_size,
                                     directory=train_dir,
                                     target_size=(408,408),#arrumar
                                     class_mode='categorical')

validation_dir = '/Users/luizeduardosantos/Documents/Turing/projeto-trainee-cv/validation'
batch_size = 1 #arrumar

val_images = idg_validation_test.flow_from_directory(batch_size=batch_size,
                                   directory=validation_dir,
                                   target_size=(408, 408),
                                   class_mode='categorical')

validation_dir = '/Users/luizeduardosantos/Documents/Turing/projeto-trainee-cv/test'
batch_size = 1 #arrumar

test_images = idg_validation_test.flow_from_directory(batch_size=batch_size,
                                   directory=validation_dir,
                                   target_size=(408, 408),
                                   class_mode='categorical')
```
