Rotação

```python
#rotação de 180 graus com for loop
rot=np.empty(np.shape(img))
linhas=np.shape(img)[0]-1
colunas=np.shape(img)[1]-1

for i in range (linhas):
    for j in range (colunas):
        rot[linhas-i,colunas-j]=img[i,j]
for j in range (colunas):
    for i in range (linhas):
        rot[linhas-i,colunas-j]=img[i,j]

rot = np.array(rot, dtype=np.uint8)
plt.imshow(rot)
plt.show()
```

```python
#rotação 180 graus com numpy
rot = np.flip(img)
plt.imshow(rot)
plt.show()
```

```python
# brilho e constraste aleatórios
alpha = np.random.random()/2 + 0.75 #contraste
beta = np.random.random()*20 #brilho

new_img = np.empty(np.shape(img))

for y in range(img.shape[0]):
    for x in range(img.shape[1]):
        for c in range(img.shape[2]):
            new_img[y,x,c] = np.clip(alpha*img[y,x,c] + beta, 0, 255)

new_img = np.array(new_img, dtype=np.uint8)
plt.imshow(new_img)
plt.show()
```

```python
#rotação, brilho, zoom aleatórios
def random_changes(image):
    linhas = np.shape(image)[0]
    colunas = np.shape(image)[1]
    
    # define o centro de rotação
    center= (np.random.randint(linhas*(24/50),linhas*(26/50)),
             np.random.randint(colunas*(24/50),colunas*(26/50)))
    #define o ângulo de rotação
    angle= (np.random.random()*360)
    #define o zoom
    scale= (np.random.random()/10)+1
    
    M = cv2.getRotationMatrix2D(center,angle,scale)
    img_rot = cv2.warpAffine(image,M,(colunas,linhas))
    return (img_rot)

img_rot = random_changes(img)

plt.imshow(img_rot)
plt.show()
```

```python
import glob
import cv2
import numpy as np
import os 
```

```python
def train_test_validation_split(arr_images, personagem):
    
    nimages = arr_images.shape[0]
    
    arr = np.arange(nimages)
    np.random.shuffle(arr)
    
    for i in range (nimages):
        if i <= nimages*0.8:
            directory = '/Users/luizeduardosantos/Documents/Turing/projeto-trainee-cv/train/'+ personagem
            os.chdir(directory)
            a = arr[i]
            image = arr_images[a]
        
            filename = 'train'+ str(i)
            cv2.imwrite(filename+'.jpeg',image)
            
            
        elif i > nimages*0.8 and i <= nimages*0.9:
            directory = '/Users/luizeduardosantos/Documents/Turing/projeto-trainee-cv/validation/'+personagem
            os.chdir(directory)
            a = arr[i]
            image = arr_images[a]
            
            filename = 'val'+ str(i)
            cv2.imwrite(filename+'.jpeg',image)
        else: 
            directory = '/Users/luizeduardosantos/Documents/Turing/projeto-trainee-cv/test/'+personagem
            os.chdir(directory)
            a = arr[i]
            image = arr_images[a]
        
            filename = 'test'+ str(i)
            cv2.imwrite(filename+'.jpeg',image)        
# talvez seja bom juntar as pastas de treino e validação porque o keras pode fazr essa divisão.
```

Todas as imagens precisam ter o mesmo tamanho, portanto definimos o tamanho de todas para (408,408) em três canais.

```python
n_images = train_images.shape[0]
resized_train_images = np.empty((n_images,408,408,3), dtype=int)

for i in range (n_images):
    resized_train_images[i] = cv2.resize(train_images[i],(408,408))

resized_train_images.shape         
```

```python
n_images = val_images.shape[0]
resized_val_images = np.empty((n_images,408,408,3), dtype=int)

for i in range (n_images):
    resized_val_images[i] = cv2.resize(val_images[i],(408,408))

resized_val_images.shape       
```

```python
n_images = test_images.shape[0]
resized_test_images = np.empty((n_images,408,408,3), dtype=int)

for i in range (n_images):
    resized_test_images[i] = cv2.resize(test_images[i],(408,408))

resized_test_images.shape         
```

Para que a imagem sirva de input para rede neural, cada uma delas precisa ser um vetor (1,408x408x3), empilhados em uma matriz

```python
vector_val_images = np.resize(resized_val_images, (183,408*408*3))
vector_val_images.shape
```

```python
vector_test_images = np.resize(resized_test_images, (183,408*408*3))
vector_test_images.shape
```

Para acelerarmos o gradiente descendente, é interessante que os valores de input sejam pequenos

```python
vector_train_images = vector_train_images/255
```

```python
vector_val_images = vector_val_images/255
```

```python
vector_test_images = vector_test_images/255
```

```python
n_images = test_images.shape[0]
test_labels = np.zeros((n_images,3), dtype=int)
d = test_images_darth.shape[0]
y = test_images_yoda.shape[0]
t = test_images_trooper.shape[0]
for i in range (d):
    test_labels[i][0] = 1
for i in range (y):
    test_labels[d + i][1] = 1
for i in range (t):
    test_labels[d + y +i][2] = 1
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
idg_train = ImageDataGenerator(rescale=1./255,
                         rotation_range=45,
                         width_shift_range=.15,
                         height_shift_range=.15,
                         horizontal_flip=True,
                         zoom_range=0.5
                        )
```

```python
idg_validation_test = ImageDataGenerator(rescale=1./255)
```

```python
path ='/Users/luizeduardosantos/Documents/Turing/projeto-trainee-cv/'
batch_size = 32 


train_dir = path + 'train'
train_images = idg_train.flow_from_directory(batch_size=batch_size,
                                             directory=train_dir,
                                             target_size=(408,408),#arrumar
                                             class_mode='categorical')

validation_dir = path + 'validation'
val_images = idg_validation_test.flow_from_directory(batch_size=batch_size,
                                                     directory=validation_dir,
                                                     target_size=(408, 408),
                                                     class_mode='categorical')

batch_size = 1
test_dir = path + 'test'
test_images = idg_validation_test.flow_from_directory(batch_size=batch_size,
                                                      directory=validation_dir,
                                                      target_size=(408, 408),
                                                      class_mode='categorical')
```

```python
model2 = keras.Sequential()
```

```python
model2.add(keras.layers.Conv2D(16, 3, padding='same', activation='relu', input_shape=(408, 408 ,3)))
model2.add(keras.layers.MaxPooling2D())

model2.add(keras.layers.Conv2D(32, 3, padding='same', activation='relu'))
model2.add(keras.layers.MaxPooling2D())

model2.add(keras.layers.Conv2D(64, 3, padding='same', activation='relu'))
model2.add(keras.layers.MaxPooling2D())

model2.add(keras.layers.Flatten())

model2.add(keras.layers.Dense(32, activation= 'relu'))
model2.add(keras.layers.Dense(8, activation= 'relu'))
model2.add(keras.layers.Dense(3, activation= 'softmax'))
```

```python
model2.summary()
```

```python
model2.compile(optimizer='adam',
              loss=tf.keras.losses.CategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])

model2.fit(train_images,epochs=20,validation_data = val_images)
```
